```python
from ultralytics import YOLO

# Load your trained model
model = YOLO('/content/forDrive/runs/detect/train3/weights/best.pt')

# Run validation on your dataset
metrics = model.val(data='/content/forDrive/data.yaml', conf=0.25, iou=0.5)

# Print overall metrics
print(f"Overall mAP@0.5: {metrics.box.map50}")
print(f"Overall mAP@0.5:0.95: {metrics.box.map}")
print(f"Overall Precision: {metrics.box.p.mean()}")
print(f"Overall Recall: {metrics.box.r.mean()}")
print(f"Overall F1 Score: {metrics.box.f1.mean()}")
```
